```python
import os
import matplotlib.pyplot as plt
import numpy as np

def plot_class_distribution(training_dir, bin_size=5):
    """
    Affiche un histogramme montrant la distribution du nombre d'images par classe dans le dossier `training`.

    :param training_dir: Chemin vers le dossier `training`.
    :param bin_size: Taille des intervalles pour les classes (par défaut : 5 images).
    """
    # Liste pour stocker le nombre d'images par classe
    num_images_per_class = []

    # Parcourir les classes dans le dossier `training`
    for class_folder in os.listdir(training_dir):
        class_path = os.path.join(training_dir, class_folder)
        if os.path.isdir(class_path):
            num_images = len(os.listdir(class_path))
            num_images_per_class.append(num_images)

    # Créer les bins pour l'histogramme
    max_images = max(num_images_per_class)
    bins = np.arange(0, max_images + bin_size, bin_size)

    # Tracer l'histogramme
    plt.figure(figsize=(10, 6))
    plt.hist(num_images_per_class, bins=bins, edgecolor='black', alpha=0.7)
    plt.title('Distribution du nombre d\'images par classe')
    plt.xlabel('Nombre d\'images par classe (par intervalle de 20)')
    plt.ylabel('Nombre de classes')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Remplacez par votre chemin vers le dossier `training`
training_dir = "archive\plantnet_300K\images_val"
plot_class_distribution(training_dir, bin_size=5)
```
